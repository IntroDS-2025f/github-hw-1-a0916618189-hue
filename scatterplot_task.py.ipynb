## 📌 題目 1：直方圖 (20 分)

從正態分布產生 1000 筆資料，畫 **bins=30** 的直方圖。

其他要求: 

> title: Normal Distribution Histogram

> xlabel: Value

> ylabel: Frequency

---

````python
import numpy as np
import matplotlib.pyplot as plt

# 1. 從正態分佈（高斯分佈）生成 1000 筆資料
# 這是確保資料分佈呈現鐘形曲線的關鍵
data = np.random.randn(1000)

# 2. 繪製直方圖
# 題目要求 bins=30，且使用 Matplotlib 預設的藍色 (即您圖片中的顏色)
plt.hist(
    data, 
    bins=30,  # 設定分箱數為 30
    edgecolor='black' # 設定邊框顏色，讓條形界線更清晰
    # 注意：這裡省略了 'color' 參數，讓它使用 Matplotlib 預設的顏色
)

# 3. 依據題目要求，設定標題和軸標籤
plt.title('Normal Distribution Histogram') # title: Normal Distribution Histogram
plt.xlabel('Value')                       # xlabel: Value
plt.ylabel('Frequency')                   # ylabel: Frequency

# 4. 確保輸出與圖片相符，不需要額外的網格線 (plt.grid() 保持註釋或移除)

# 5. 將圖表儲存為檔案 (這一步非常重要，以便上傳到 GitHub)
output_filename = 'normal_distribution_histogram.png'
plt.savefig(output_filename) 
print(f"圖表已儲存為：{output_filename}")

# 6. (可選) 顯示圖表
# plt.show()
````

````python
# 你的程式

import numpy as np
import matplotlib.pyplot as plt

# 設定隨機種子，確保每次執行的結果（隨機數）都一樣
np.random.seed(768)

# 設定資料點的數量
n = 200

# 1. 生成 x 軸資料：從平均值為 2.0，標準差為 1.0 的正態分佈中生成 n 個點
x = np.random.normal(loc=2.0, scale=1.0, size=n)

# 2. 生成 y 軸資料：y 與 x 相關，並加入一些噪音 (標準差 0.8 的正態分佈)
y = x * 0.5 + np.random.normal(scale=0.8, size=n)

# 3. 生成點的大小變量 (size_var)：從平均值為 20，標準差為 10 的正態分佈中生成，並取絕對值
size_var = np.abs(np.random.normal(loc=20, scale=10, size=n))

# 4. 將點的大小進行縮放
size_scale = size_var * 5

# 5. 繪製散佈圖 (Scatter Plot)
scatter = plt.scatter(
    x, y,
    c=y,                 # 根據 y 值來決定點的顏色 (Color)
    s=size_scale,        # 根據 size_scale 來決定點的大小 (Size)
    cmap='viridis',      # 使用 'viridis' 顏色映射 (Colormap)
    alpha=0.7            # 設定點的透明度
)

# 6. 設定圖表標題
plt.title('Scatter with Colormap')

# 7. 繪製色條 (Color Bar)，並設定其標籤
cbar = plt.colorbar(scatter)
cbar.set_label('y-values')

# 8. (重要) 將圖表儲存為檔案，以便上傳到 GitHub
output_filename = 'scatterplot_with_colormap.png'
plt.savefig(output_filename)
print(f"圖表已儲存為：{output_filename}")

# 9. 顯示圖表
plt.show()

````
