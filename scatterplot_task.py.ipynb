````python
# ‰Ω†ÁöÑÁ®ãÂºè

import numpy as np
import matplotlib.pyplot as plt

# Ë®≠ÂÆöÈö®Ê©üÁ®ÆÂ≠êÔºåÁ¢∫‰øùÊØèÊ¨°Âü∑Ë°åÁöÑÁµêÊûúÔºàÈö®Ê©üÊï∏ÔºâÈÉΩ‰∏ÄÊ®£
np.random.seed(768)

# Ë®≠ÂÆöË≥áÊñôÈªûÁöÑÊï∏Èáè
n = 200

# 1. ÁîüÊàê x Ëª∏Ë≥áÊñôÔºöÂæûÂπ≥ÂùáÂÄºÁÇ∫ 2.0ÔºåÊ®ôÊ∫ñÂ∑ÆÁÇ∫ 1.0 ÁöÑÊ≠£ÊÖãÂàÜ‰Ωà‰∏≠ÁîüÊàê n ÂÄãÈªû
x = np.random.normal(loc=2.0, scale=1.0, size=n)

# 2. ÁîüÊàê y Ëª∏Ë≥áÊñôÔºöy Ëàá x Áõ∏ÈóúÔºå‰∏¶Âä†ÂÖ•‰∏Ä‰∫õÂô™Èü≥ (Ê®ôÊ∫ñÂ∑Æ 0.8 ÁöÑÊ≠£ÊÖãÂàÜ‰Ωà)
y = x * 0.5 + np.random.normal(scale=0.8, size=n)

# 3. ÁîüÊàêÈªûÁöÑÂ§ßÂ∞èËÆäÈáè (size_var)ÔºöÂæûÂπ≥ÂùáÂÄºÁÇ∫ 20ÔºåÊ®ôÊ∫ñÂ∑ÆÁÇ∫ 10 ÁöÑÊ≠£ÊÖãÂàÜ‰Ωà‰∏≠ÁîüÊàêÔºå‰∏¶ÂèñÁµïÂ∞çÂÄº
size_var = np.abs(np.random.normal(loc=20, scale=10, size=n))

# 4. Â∞áÈªûÁöÑÂ§ßÂ∞èÈÄ≤Ë°åÁ∏ÆÊîæ
size_scale = size_var * 5

# 5. Áπ™Ë£ΩÊï£‰ΩàÂúñ (Scatter Plot)
scatter = plt.scatter(
    x, y,
    c=y,                 # Ê†πÊìö y ÂÄº‰æÜÊ±∫ÂÆöÈªûÁöÑÈ°èËâ≤ (Color)
    s=size_scale,        # Ê†πÊìö size_scale ‰æÜÊ±∫ÂÆöÈªûÁöÑÂ§ßÂ∞è (Size)
    cmap='viridis',      # ‰ΩøÁî® 'viridis' È°èËâ≤Êò†Â∞Ñ (Colormap)
    alpha=0.7            # Ë®≠ÂÆöÈªûÁöÑÈÄèÊòéÂ∫¶
)

# 6. Ë®≠ÂÆöÂúñË°®Ê®ôÈ°å
plt.title('Scatter with Colormap')

# 7. Áπ™Ë£ΩËâ≤Ê¢ù (Color Bar)Ôºå‰∏¶Ë®≠ÂÆöÂÖ∂Ê®ôÁ±§
cbar = plt.colorbar(scatter)
cbar.set_label('y-values')

# 8. (ÈáçË¶Å) Â∞áÂúñË°®ÂÑ≤Â≠òÁÇ∫Ê™îÊ°àÔºå‰ª•‰æø‰∏äÂÇ≥Âà∞ GitHub
output_filename = 'scatterplot_with_colormap.png'
plt.savefig(output_filename)
print(f"ÂúñË°®Â∑≤ÂÑ≤Â≠òÁÇ∫Ôºö{output_filename}")

# 9. È°ØÁ§∫ÂúñË°®
plt.show()

````
